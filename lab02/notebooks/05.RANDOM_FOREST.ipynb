## 05. Random Forest

Random Forest, en este caso, es un método de clasificación que utiliza un conjunto de muchos árboles de decisión diferentes. En este método, la decisión de clasificar en una clase u otra proviene de hacer la media de los resultados de todos los árboles de decisión. Random Forest tiene una pequeña variación y es difícil tener un sobreajuste. Usamos este método porque un árbol de decisión único es muy inestable y, por esa razón, entrenamos el modelo con muchos árboles de decisión.

### Creamos el Modelo de Random Forest

Ahora cargamos las variables de las 101 columnas de entrada en X excluyendo la columna “estado_prestamo” con el método drop(). En cambio agregamos la columna “estado_prestamo” en la variable y. Ejecutamos X.shape para comprobar la dimensión de nuestra matriz con datos de entrada de 475 registros por 101 columnas.

```python
X = prestamo.drop('estado_prestamo', axis = 1)
y = prestamo['estado_prestamo']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12345)
```

Y creamos nuestro modelo y hacemos que se ajuste (fit) a nuestro conjunto de entradas X y salidas ‘y’.

```python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

Y confirmamos cuan bueno fue nuestro modelo utilizando model.score() que nos devuelve la precisión media de las predicciones, en nuestro caso del 92,63%.

```python
classifier.score(X_test, y_test)
```

Si entrenamos el modelo con 100.000 variables, la precisión del modelo, baja a 91,73%.
