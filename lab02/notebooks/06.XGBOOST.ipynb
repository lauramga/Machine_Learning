## 06. XGBoost

XGBoost es un método de clasificación similar al modelo Random Forest, que utiliza un conjunto de muchos árboles de decisión diferentes. La diferencia con Random Forest es que XGBoost tiene más varianza, podría tener problemas con el sobreajuste y generalmente tiene más precisión con menos estimadores. En este método los casos que han fallado previamente en su clasificación tienen más importancia para el modelo, por lo que se vuelven a utilizar en las próximas muestras para intentar clasificarlos correctamente.

```python
# Libraries

import numpy as np
import pandas as pd

# Plots

import matplotlib
from matplotlib import pyplot as plt # plots
import seaborn as sns # plots

# Metricas

from scipy import stats as sts # normality test JB
from sklearn import model_selection, metrics
from sklearn.model_selection import train_test_split # split data
from sklearn.metrics import accuracy_score, mean_squared_error # MSE
from sklearn.preprocessing import StandardScaler # estandarization
from sklearn import preprocessing # estandarization
from sklearn.svm import SVC # SVM
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score, plot_roc_curve
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_classification
from math import sqrt # sqrt
import itertools # aplanar arrays
import xgboost as xgb
from xgboost.sklearn import XGBClassifier, XGBRegressor
import math
from imblearn.over_sampling import SMOTE

import random
random.seed(12345)

from sklearn.impute import KNNImputer
import pickle
```

```python
de_set = '../data/DE_set.csv'
```

```python
prestamo = pd.read_csv(de_set)
```

```python
prestamo.head(10)
```

```python
X = prestamo.drop(['estado_prestamo'], axis = 1).values
y = prestamo['estado_prestamo'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =12345)
```

```python
%%time
estimator = xgb.XGBClassifier(n_estimator=100, subsample=0.8,
                             colsample_bytree=0.8, objective= 'binary:logistic',
                             scale_pos_weight=1, seed=40, nthread = 6)
estimator.fit(X_train, y_train)
```

```python
pred = estimator.predict(X_test)
```

```python
accuracy_score(y_test, pred)
```

Si entrenamos el modelo con 100.000 variables, la precisión del modelo, baja a 91,57%.
